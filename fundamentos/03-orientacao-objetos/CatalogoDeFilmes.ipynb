```kotlin
data class Filme(
    val titulo: String,
    val genero: String,
    val classificacaoIndicativa: String,
    val anoLancamento: Int,
)

class Catalogo(
    protected var filmes: MutableList<Filme> = mutableListOf()
){
    fun adicionarFilme(filme: Filme):Unit {
        this.filmes.add(filme)
        println("Filme: ${filme.titulo} Adicionado com sucesso")
    }

    fun buscarPorGenero(genero: String): Unit {
        val filtro = this.filmes.filter { it.genero.contains(genero, ignoreCase = true) }

        if (filtro.isEmpty())
        {
            println("NÃ£o hÃ¡ nenhum filme com esse genero")
            return
        }
        println("=====================================================================================================================")
        println("| Filmes por gÃªneros: ${genero} | Quantidade: ${filtro.size}")
        println("=====================================================================================================================")
        filtro.forEach {
            println("| TÃ­tulo: ${it.titulo} | GÃªnero: ${it.genero} | ClassificaÃ§Ã£o: ${it.classificacaoIndicativa} | Ano: ${it.anoLancamento}")
        }
        println("=====================================================================================================================")
    }

    fun listarFilmes():Unit {
        println("=====================================================================================================================")
        println("| Todos os Filmes no CatÃ¡logo")
        println("=====================================================================================================================")
        this.filmes.forEach {
            println("| TÃ­tulo: ${it.titulo} | GÃªnero: ${it.genero} | ClassificaÃ§Ã£o: ${it.classificacaoIndicativa} | Ano: ${it.anoLancamento}")
        }
        println("=====================================================================================================================")
    }

    fun filmesPorAno(ano:Int):Unit {
        val filtro = this.filmes.filter { it.anoLancamento.equals(ano) }
        println("=====================================================================================================================")
        println("| Filmes por Ano : ${ano} | Quantidade: ${filtro.size}")
        println("=====================================================================================================================")
        filtro.forEach {
            println("| TÃ­tulo: ${it.titulo} | GÃªnero: ${it.genero} | ClassificaÃ§Ã£o: ${it.classificacaoIndicativa} | Ano: ${it.anoLancamento}")
        }
        println("=====================================================================================================================")
    }

}

var Filme1 = Filme("XMEN","aÃ§Ã£o","10anos",2007)
var Filme2 = Filme("JohnWick","Comedia","14anos",2007)
var Filme3 = Filme("PulpFiction","Comedia","16anos",2008)
var Filme4 = Filme("Velozes e furiosos","Comedia","18anos",2008)


var Catalogo = Catalogo()

Catalogo.adicionarFilme(Filme1)
Catalogo.adicionarFilme(Filme2)
Catalogo.adicionarFilme(Filme3)
Catalogo.adicionarFilme(Filme4)

Catalogo.buscarPorGenero("comedia")
Catalogo.listarFilmes()
Catalogo.filmesPorAno(2008)

```
